```python
import config
from sqlalchemy import create_engine

eng = create_engine(config.DB_URL)
conn = eng.connect()
```

```python
pd.options.display.max_rows = 1000
```

```python
sql1 = '''SELECT
   s.site_id,
   s.name,
   s.img_url,
   s.elev_ft,
   s.capacity,
   ROUND(s.capacity * 1.0 - t1.spots,1) AS taken_avg_overall,
   s.capacity * 1.0 - t6.spots AS taken_min,
   ROUND(s.capacity * 1.0 - t2.spots,1) AS taken_avg_range,
   s.capacity * 1.0 - t3.spots AS taken_10_28,
   s.capacity * 1.0 - t4.spots AS taken_10_29,
   s.capacity * 1.0 - t5.spots AS taken_10_30
   
   FROM
(SELECT site_id,
       AVG(spots * 1.0) AS spots
  FROM res_date_min GROUP BY site_id) as t1
   LEFT JOIN
   (SELECT site_id,
       AVG(spots * 1.0) AS spots
  FROM res WHERE scrape_date = (SELECT MAX(scrape_date) FROM res) AND res_date BETWEEN '2021-10-28' AND '2021-10-30' GROUP BY site_id) as t2
  ON t1.site_id = t2.site_id
  LEFT JOIN
   (SELECT site_id,
       spots * 1.0 AS spots
  FROM res WHERE scrape_date = (SELECT MAX(scrape_date) FROM res) AND res_date = '2021-10-28') as t3
  ON t1.site_id = t3.site_id
  LEFT JOIN
   (SELECT site_id,
       spots * 1.0 AS spots
  FROM res WHERE scrape_date = (SELECT MAX(scrape_date) FROM res) AND res_date = '2021-10-29') as t4
  ON t1.site_id = t4.site_id
  LEFT JOIN
   (SELECT site_id,
       spots * 1.0 AS spots
  FROM res WHERE scrape_date = (SELECT MAX(scrape_date) FROM res) AND res_date = '2021-10-30') as t5
  ON t1.site_id = t5.site_id
  LEFT JOIN
  (SELECT site_id,
       spots * 1.0 AS spots
  FROM res_min) as t6
   ON t1.site_id = t6.site_id
  LEFT JOIN
  sites s ON t1.site_id = s.site_id
  ORDER BY s.site_id;'''
```

```python
import pandas as pd
```

```python
df = pd.read_sql(sql1, conn)
```

```python
df[(df['capacity'] <= 8)].sort_values(by=['taken_min', 'taken_avg_range'])
```

```python
df[(df['taken_avg_range'] <= 8)].sort_values(by=['taken_min', 'taken_avg_range'])
```
